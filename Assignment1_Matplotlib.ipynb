```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.random.randn(1000)
x2=np.random.randn(1000)
y=np.random.randn(1000)

plt.scatter(y,x1,marker='o',c='b')
plt.scatter(y,x2,marker='v',c='y')
plt.title('Scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
import matplotlib.pyplot as plt


Country=['USA','BRAZIL','RUSSIA','INDIA','UK']
Total_Case=[2779953,1453369,654405,605220,313483]
colours=['r','b','g','c','y']
plt.bar(Country,Total_Case,color=colours,width=0.4)
plt.title('Corana Virus Cases')
plt.xlabel('Countries')
plt.ylabel('Cases')

plt.show()
```
