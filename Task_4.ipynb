###### Department wise codekata performence (pie chart)

```python
a=df['Department'].unique()
a
```

```python
cse=df[df['Department']=='Computer Science and Engineering']
a=cse['CodeKata Score']
c=sum(a)
```

```python
ecee=df[df['Department']=='Electronics and Communication Engineering']
a=ecee['CodeKata Score']
e=sum(a)
```

```python
data=[c,e,ee]
a=df['Department'].unique()
labels=a

colors = sns.color_palette('pastel')[0:5]
plt.pie(data,labels=labels,colors = colors, autopct='%.0f%%')
plt.show()
```
