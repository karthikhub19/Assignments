```python
###Generate sine and cosine plot over a time scale of 0 to 5 with an interval of 0.5. 
###Plot them on two different subplots (12x) and apply xlabel, ylabel, Grids as well as the title.
```

```python
from matplotlib import figure
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,5,0.5)
y1=np.sin(x)
y2=np.cos(x)


plt.subplot(1,2,1)
plt.plot(x,y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.title('sine wave')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.title('cose wave')

plt.show()
```
