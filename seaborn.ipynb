```python
import seaborn as sns
df=sns.load_dataset('penguins')
print(df)
sns.pairplot(df,hue='species',markers=['o','s','d'])
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
print(df)
sns.boxplot(x='sex',y='tip',hue='smoker',data=df)
```
